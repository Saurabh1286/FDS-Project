```python
Temp = Type
plt.figure(figsize=(20,10))
x = Temp["States/UTs"].iloc[:-1]
xn = np.arange(x.count())
bar_width = 0.2
y = Temp["Total Accidents"].iloc[:-1]
y1 = Temp["Fatal Accidents"].iloc[:-1]
y2 = Temp["Grievous Injury Accidents"].iloc[:-1]
y3 = Temp["Minor Injury Accidents"].iloc[:-1]
y4 = Temp["Non-Injury Accidents"].iloc[:-1]
plt.bar(xn - 2 * bar_width, y1,bar_width, color = "Red", align = "center", label = "Fatal")
plt.bar(xn - 1 * bar_width, y2,bar_width, color = "Blue", align = "center", label = "Grievous Injury")
plt.bar(xn - 0 * bar_width, y3,bar_width, color = "Green", align = "center", label = "Minor Injury")
plt.bar(xn + 1 * bar_width, y4, bar_width, color = "Orange", align = "center", label = "Non-Injury")
#plt.bar(xn + 2 * bar_width, y, bar_width, facecolor = "Black", align = "center", label = "Total")
plt.legend()
plt.xticks(xn + 0.2,x,  rotation=90)
plt.xlabel("States/UTs")
plt.ylabel("Number of Accidents")
plt.title("States/UTs wise Number of Accidents based on Type of Accident")
plt.show()
```
